```python
# Country code
country_df =pd.read_excel('../input/zomato-restaurants-data/Country-Code.xlsx')

# combining country and zomato dataframe

final_df = pd.merge(zomato_df, country_df,on='Country Code', how='left',)
final_df.head()
```

## Number of Restorunts per country

```python
final_df.Country.value_counts()
country_names = final_df.Country.value_counts().index
country_counter = final_df.Country.value_counts().values
```

### list of top3 countries using zomato and how many restorunts those countries have

```python

plt.pie(x=country_counter[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Zomato has largest customer base in India with 94.39% followed by USA 4.73% and United Kingdom 0.87%.

### Ratings, Rating Text and Aggregated Ratings Analysis

```python
ratings_df=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings_df
```

### Observation
1. Rating >= 4.5  => **Excellent**
2. Rating >= 4.0 and Rating <= 4.4 =>**Very Good**
3. Rating >= 3.5 and Rating <= 3.9 => **Good**
4. Rating >= 2.0 and Rating <=3.4 => **Average**
5. Rating >= 0.0 and Rating <= 1.99 => **Poor**
6. Rating == 0.0 => No Rating (**UnRated**)

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',
            y='Rating Count',
            data=ratings_df,hue='Rating color',
            palette=['Blue', 'Red', 'Orange','Yellow','Green','Green'],
           )
```

```python
sns.countplot(x= 'Rating color', data=ratings_df,hue='Rating color',
            palette=['Blue', 'Red', 'Orange','Yellow','Green','Green'])
```
